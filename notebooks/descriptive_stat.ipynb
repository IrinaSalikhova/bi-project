```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
df : pd.DataFrame = pd.read_csv("../data/cleaned.csv")

print(df.columns)
df.head()
```

```python
# Select only numeric columns
column_names = ['amount', 'card_limit', 'user_income', 'user_debt', 'user_rating', 'user_number_of_cards', 'user_age']
numeric_columns = df[column_names]

# Calculate Mean for all numeric columns
mean_values = numeric_columns.mean()

# Calculate Median for all numeric columns
median_values = numeric_columns.median()

# Calculate Mode for all numeric columns
mode_values = numeric_columns.mode().iloc[0]  # Mode returns a DataFrame, so we take the first row

# Display the results
for col in numeric_columns.columns:
    print(f"Column: {col}")
    print(f"Mean: {mean_values[col]}")
    print(f"Median: {median_values[col]}")
    print(f"Mode: {mode_values[col]}")
    print("\n")
```

```python
# Calculate Range for selected columns
range_values = numeric_columns.max() - numeric_columns.min()

# Calculate Variance for selected columns
variance_values = numeric_columns.var()

# Calculate Standard Deviation for selected columns
std_dev_values = numeric_columns.std()

# Display the results
for col in numeric_columns:
    print(f"Column: {col}")
    print(f"Range: {range_values[col]}")
    print(f"Variance: {variance_values[col]}")
    print(f"Standard Deviation: {std_dev_values[col]}")
    print("\n")
```

```python
# find outliners using z-score 
from scipy import stats

z_scores = stats.zscore(numeric_columns)
abs_z_scores = abs(z_scores)

outliers = (abs_z_scores > 3).all(axis=1)

# Display rows containing outliers
print(df[outliers])

# we don't have outliers since data cleaning was done before
```

```python
# Select only numeric columns
column_names = ['amount', 'card_limit', 'user_income', 'user_debt', 'user_rating', 'user_number_of_cards', 'user_age', 'flagged_as_fraud']
numeric_columns = df[column_names]

# Plot a pair plot for the tips dataset. Test the hue (by flagged_as_fraud) argument
sns.pairplot(data = numeric_columns, hue='flagged_as_fraud')
```

```python
# Select only amount, user income, flagged_s_fraud columns
column_names = ['amount', 'user_income', 'flagged_as_fraud']
numeric_columns = df[column_names]

# Plot a plot for only amount, user income, flagged_s_fraud columns. Test the hue (by flagged_as_fraud) argument
plt.scatter(numeric_columns['amount'].where(-numeric_columns['flagged_as_fraud']), numeric_columns['user_income'].where(-numeric_columns['flagged_as_fraud']), color='lightgreen')
plt.scatter(numeric_columns['amount'].where(numeric_columns['flagged_as_fraud']), numeric_columns['user_income'].where(numeric_columns['flagged_as_fraud']),color='red')
plt.xlabel('amount')
plt.ylabel('income')
plt.title('amount vs income')
plt.legend(['Not fraud','Fraud'], loc='lower right')
plt.show()
```

```python
# Select only numeric columns
column_names = ['amount', 'flagged_as_fraud', 'card_limit', 
      'user_birth_year', 'user_birth_month', 'user_income', 'user_debt', 'user_rating',
      'user_number_of_cards', 'user_age', 'state_match', 'city_match', 'error_count', 'is_error', 'bad_cvv',
      'bad_card_number', 'bad_expir', 'bad_pin', 'bad_zip', 'insuf_balance',
      'tech_glitch', 'retirement_age', 'is_retired', 'is_card_expired', 'log_amount']

numeric_columns = df[column_names]

# Plot a heatmap of the correlation matrix.
correlations = numeric_columns.corr()
correlations
```
